Sentence Tokenization

```python
import regex as re 
def sentence_tokenize(text):
    # regex to split senetences by periods followed by a space , handling abbreviations and dates 
    # abbrevations and accronmys where it is not treated as sentence boundary . 
    # abbreviation means a word that is shortened form of a longer word or phrase . eg. Dr. for Doctor , Inc. for Incorporated .and accronym is a word formed from the initial letters of a phrase . eg. NASA for National Aeronautics and Space Administration .
    # regex is used to split sentences by periods followed by a space , but it also checks for common abbreviations and dates to avoid splitting sentences incorrectly .
    # example 1 : "Dr. Smith is here. He will meet you at noon."
    # in this example , the regex will correctly identify "Dr." as an abbreviation and will not split the sentence at that point . it will split the sentence at the period after "here" and "noon" , resulting in two sentences : "Dr. Smith is here." and "He will meet you at noon."
    abbreviations= r'\b(?:Dr|Mr|Ms|Mrs|Sr|Jr|Prof|St|Rev|Dec|Nov|Jan|Feb|Mar|Apr|Aug|Sep|Oct|etc)\.'
    # regax to split on period not preceded by an abbreviation
    sentences=re.split(r'(?<!' +abbreviations +r')(?<!\b[A-Z][a-z]{0,3})\.\s+', text)
    return[sentence.strip()+ '.' for sentence in sentences if sentence ]

# example usage
text= "Dr.smith went to washington. On Dec.25, 2024, the event will occur. He said,'i'll be there.'"
sentences=sentence_tokenize(text)
print("sentence Tokenization:",sentences)

```

**The above code perform TextPreprocessing using regax libaray**


*It read the paragraph and detect the sentence boundaries*

*it avoid breaking after abbreviations and split sentences correctly*

*then it return a list of sentences*    

**Word Tokenization**

```python
# regex  is a regular expresion used for searching ,finding, substituting , marching the text especially from the text.
# from the above text that is split into sentences we will split into tokens. 
def word_tokenizer(text):
    #reg Regular expression to match words
    #this pattern matches sequence of alphanumeric characters and apostrophes, treating them as words. 
    #words=re.findall(r"\b\w+(?:'\w+)?\b",text)
    words = re.findall(r"\b\w+(?:'\w+)?\b|[.,!?;]", text) # this pattern matches words as well as punctuation marks such as periods, commas, exclamation points, question marks, and semicolons. and it treats them as separate tokens.
    return words

# giving the previous sentences list as input to the word tokenizer
for sentence in sentences:
    print("sentence input:%s"%(sentence)) # this line is used to print the sentence that is being tokenized into words. it uses string formatting to insert the sentence into the output string.
    words=word_tokenizer(sentence) 
    print("word list:",end="") 
    print(words)
```
