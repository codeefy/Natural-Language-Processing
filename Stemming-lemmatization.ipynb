**Demo2-Stemming-Lemmatization**

```python
def simple_stem(word):
    suffixes=['ing','ed','ly','es','s','ment',]
    for suffix in suffixes:
        if word.endswith(suffix):
# this line removes the suffix from the word by slicing it up to the length of the suffixa and returns the stemmed word. for example, if the word is "running" and the suffix is "ing", it will return "runn".
            return word[:-len(suffix)] 
        return word

    #example usage 
words= ['running','studied','cares','happiness']
stems=[simple_stem(word) for word in words] 
print(stems)
```
