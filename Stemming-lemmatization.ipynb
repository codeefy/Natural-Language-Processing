```python
def simple_stem(word):
    suffixes=['ing','ed','ly','es','s','ment','ness']
    #suffixes=['ing','ed','ness','ly','es','s','ment']
    for suffix in suffixes:
        if word.endswith(suffix):
# this line removes the suffix from the word by slicing it up to the length of the suffixa and returns the stemmed word. for example, if the word is "running" and the suffix is "ing", it will return "runn".
            return word[:-len(suffix)] 
    return word

    #example usage 
words= ['running','studied','cares','happiness']
stems=[simple_stem(word) for word in words] 
print(stems)
# if we change the position of 'ness' in the suffixes list to be before 's', it will stem "happiness" to "happines" instead of "happines". Therefore, the order of suffixes in the list can affect the stemming results.
```

**Lemmatization**

```python
def simple_lemmatize(word):
    irregulars={'better':'good','worse':'bad','caring':'care','studies':'study','running':'run'}
    if word in irregulars:
        return irregulars[word]
    suffixes={'ing':'','ed':'','ly':'','es':'','s':'','ment':'','ness':''}
    for suffix in suffixes:
        if word.endswith(suffix):
            return word[:-len(suffix)]+suffixes[suffix] # this line removes the suffix from the word by slicing it up to the length of the suffix and then adds the corresponding lemma ending from the suffixes dictionary. for example, if the word is "running" and the suffix is "ing", it will return "run" by removing "ing" and adding "" (empty string) from the suffixes dictionary.
    return word
#example usage 
words=['running','studies','better','caring']
lemmas=[simple_lemmatize(word) for word in words]
print(lemmas)

```
